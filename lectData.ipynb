```python
### s√≠ntasis de 'loc' :
### [row inicial:row final, col inicial:col final]
# nombres de las columnas y renglones
```

```python
ndf = df.loc[:, :'Glasses']
ndf.head(2)
```

```python
# hay que recordar que para llamar una lista completa es [:] (de principio a fin)
```

```python
### s√≠ntasis de 'loc' :
### [row inicial:row final, col inicial:col final]
# √≠ndices de las columnas y renglones
```

```python
ndf = df.iloc[3:5, :5]
ndf.head(2)
```
