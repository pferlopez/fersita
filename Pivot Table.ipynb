# Pivot Table

```python
import pandas as pd
import numpy as np
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/census.csv')
df.head(2)
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'STNAME'
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'STNAME',
                          values = 'BIRTHS2010'
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'STNAME',
                          values = 'BIRTHS2010',
                          aggfunc = 'sum'
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'STNAME',
                          values = 'BIRTHS2010',
                          aggfunc = ['mean','sum','median'],
                          columns = 'SUMLEV'
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'STNAME',
                          values = 'BIRTHS2010',
                          aggfunc = ['mean','sum','median'],
                          margins = True
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'STNAME',
                          values = 'BIRTHS2010',
                          aggfunc = 'median',
                          columns = 'REGION',
                          margins = True
                         )
my_pivot
```
