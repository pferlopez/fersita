```python
import pandas as pd
```

```python
df = pd.read_excel('datasets/Canada.xlsx',
                   engine = 'openpyxl',
                   sheet_name = 'Canada by Citizenship',
                  skiprows = range(20)
                  )
df
```

```python
df.columns
```

```python
# limpieza de database, se eliminan algunas columnas
```

```python
df.drop(['Type',    'Coverage',      'REG',     'DEV',
           'Unnamed: 43',
       'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47',
       'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'],
      axis = 1,
        inplace = True)
```

```python
df.columns
```

```python
df.rename(columns = {'OdName':'Country',
                   'AreaName':'Continent',
                   'RegName':'Region'},
        inplace=True)
df.columns
```

```python
df.head(2)
```

```python
# se agrega una columna de total tomando en cuenta los años como indexes
```

```python
df['Total']=df.iloc[:,5:].sum(axis=1)
df.head(2)
```

```python
# se agrega una columna de total tomando en cuenta los años como strings
```

```python
df.columns=list(map(str,df.columns))
df.columns
```

```python
df['Total2']=df.loc[:,'1980':'2013'].sum(axis=1)
df.head(2)
```
