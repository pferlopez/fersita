# A3 Pivot Tables

```python
import pandas as pd
import numpy as np
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'desc_ent',
                          values = 'peso',
                          aggfunc = 'mean',
                          margins = True
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['desc_ent','sexo'],
                          values = 'peso',
                          aggfunc = 'mean',
                          margins = True
                         )
my_pivot
```
