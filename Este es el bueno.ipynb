# Descripción de las variables de interés

* name - nombre de la canción, variable categórica
* album - nombre del album, variable categórica
* artist - nombre del artista o artistas involucrados, variable categórica
* lenght - largo de la canción en milisegundos, variable numérica
* popularity - porcentaje de popularidad de la canción basada en el algoritmo de Spotify, variable numérica
* danceability - qué tan adecuada es una pista para bailar en función de una combinación de elementos musicales que incluyen el tempo, la estabilidad del ritmo, la fuerza del ritmo y la regularidad general, variable numérica
* acousticness - qué tan acústica es la canción, variable numérica
* energy - una medida perceptiva de intensidad y actividad, variable numérica
* valence - medida de qué tan feliz es la canción, variable numérica

# Pregunta de investigación 2:
## ¿Lo bailable que se considera una canción depende del tempo en el que esta esté?

Sabemos que la variable tempo representa los beats por minuto de la canción, por lo que usualmente creeríamos que mientras mayor sea el tempo mayor es el porcentaje de baileabilidad.

```python
sns.scatterplot(x='danceability', y='tempo', data=df, hue='album')
sns.set(rc={'figure.figsize':(20,11)})
plt.xlabel('danceability',fontsize=17)
plt.ylabel('tempo',fontsize=17)
plt.title('danceability vs tempo',fontsize=17)
plt.show()
```

# Pregunta de investigación 5:
## ¿Cuál album tiene menos canciones y cuál tiene más?

```python
counts = df['album'].value_counts()
counts = counts.to_frame()
counts

counts['%'] = (counts.iloc[:,:]/counts.loc[:,'album'].sum(axis=0))*100
counts.reset_index(inplace=True)
counts = counts.rename({'index':'album','album':'# de canciones'}, axis=1)
counts
```

```python
colores = sns.color_palette("flare")
counts['# de canciones'].plot(kind='pie',
                              shadow=True,
                              autopct='%1.1f%%',
                              pctdistance=1.15,
                              colors = colores,
                             labels=None)
plt.legend(labels=counts.album,
          bbox_to_anchor=(1,1))
plt.title('número de canciones por album', fontsize=20)
```

## Conclusión: 
El album que tiene más canciones es el de "Fearless (Taylor's Version)" con 26 canciones (15.2% del total de canciones de Taylor Swift) y los álbumes con menos canciones son "Taylor Swift" y "reputation" con 15 canciones (8.8% del total de canciones de Taylor Swift).

# Pregunta de investigación 6:
## ¿Cuál sería el impacto de energy en popularity?

Modelo de regresión:

popularity = b0 + b1*energy

* b1 es el impacto de energy en popularity
* b0 es la popularity cuando energy no tiene impacto en popularity

Variable dependiente: popularity
Variable independiente: energy

```python
df.corr()
```

### Al observar esta tabla de correlación podemos ver que popularity y energy tienen una correlación positiva

```python
modelA = sm.OLS.from_formula('popularity~energy', data=df)
resultA = modelA.fit()
resultA.summary()
```

* Nótese que la R cuadrada del modelo es muy baja, por lo que el 1.6% de popularity está descrito por energy.

b0 = 56.5367

b1 = 8.0061

### El modelo final sería:
popularity = 56.5367 + 8.0061*energy

```python
print('La popularidad de una canción con un porcentaje de energy de 0.491 es : \t',56.5367+8.0061*0.491)
print('Su residual sería :\t',49-60.4676951)
```
